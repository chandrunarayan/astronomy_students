```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(100)  ## the arrange method simply defines an interval between points for a given range
y = np.sin(x)
plt.plot(x,y)
plt.show()
## x and y are arrays!!
x
```

```python
x = np.arange(0,100, np.pi/4) ## the linspace method defines the actual number points within the range
y = np.sin(x)
plt.plot(x,y)
plt.show()
#x
```

```python
x = np.linspace(0, 4*np.pi, num=1000) ## the linspace method defines the actual number points within the range
y = np.sin(x)
plt.plot(x, y)
plt.title("look!!! it's sine!!!!")
plt.xlabel("Radians")
plt.ylabel("sin(x)")
plt.show()
```

```python
def deg2rad(x):
    return x * np.pi / 180
x = np.linspace(0,720, num=1000) ## the linspace method defines the actual number points within the range
y = np.sin(deg2rad(x))
plt.plot(x,y)
plt.title("sine curve but degree!!11!!!")
plt.xlabel("Degrees")
plt.ylabel("sin(x)");
plt.show()
```

```python
rad = 5

def yval(x):
    return np.sin(x) * rad

def xval(y):
    return np.cos(y) * rad

x = np.linspace(0,2*np.pi, num=1000) ## the linspace method defines the actual number points within the range
y = np.linspace(0,2*np.pi, num=1000)
plt.plot(xval(y),yval(x))
plt.title("A parametric Circle of XY with a curved line connecting them!")
plt.xlabel("xval")
plt.ylabel("yval");
plt.gca().set_aspect('equal')
plt.text(0, 0, "+")
plt.text(0.3, -0.3, "(0,0)")
plt.show()
```

```python
rad = 5
u = 3
v = -4

def yval(x):
    return np.sin(x) * rad

def xval(y):
    return np.cos(y) * rad

x = np.linspace(0,2*np.pi, num=100) ## the linspace method defines the actual number points within the range
y = np.linspace(0,2*np.pi, num=100)
plt.plot(u+xval(y),v+yval(x), color="red")    # plot line
plt.scatter(u+xval(y),v+yval(x), color="red") # plot dots
plt.scatter(u,v,marker='+')  # plot center point
plt.text(2.75,-3.25,"radius: %1d" % rad)
plt.text(u+0.05,v-0.05,"center: (%1.1f, %1.1f)" % (u,v))
plt.title("An offset Circle with line and dots!")
plt.xlabel("xval")
plt.ylabel("yval")
plt.gca().set_aspect('equal')
x = x[75:]
y = y[75:]
plt.scatter(u+xval(y),v+yval(x), color="blue")
plt.show()
```
